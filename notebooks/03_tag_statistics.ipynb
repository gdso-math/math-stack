```python
import pandas as pd
```

```python
chemistry = pd.read_csv("../data/csv/chemistry_post_index_tags.csv", lineterminator='\n')
chemistry.head()
```

```python
chemistry.info()
```

```python
chemistry.dropna(inplace = True)
chemistry.reset_index(drop = True, inplace = True)
chemistry
```

```python
import re
```

```python
pattern = re.compile("<(.+?)>")
chemistry['Tag_list'] = chemistry.apply(lambda row: re.findall(pattern, row.Tags), axis=1)
```

```python
chemistry.head()
```

```python
filter_chemistry = chemistry[chemistry.apply(lambda row: len(row.Tag_list) > 1, axis=1)]
filter_chemistry.reset_index(drop = True, inplace = True)
filter_chemistry.head()
```

```python
from itertools import combinations
filter_chemistry['Tag_combination'] = filter_chemistry.apply(lambda row: list(combinations(row.Tag_list, 2)), axis = 1) 
filter_chemistry.head()
```

```python
filter_chemistry.shape
```

```python
def tag_comb_freq(subject):
    subj_dict = {}
    for tags in subject["Tag_combination"]:        
        for tag in tags:
            tag = tuple(sorted(list(tag)))
            if tag not in subj_dict:
                subj_dict[tag] = 1
            else:
                subj_dict[tag] += 1
    return subj_dict
```

```python
import operator
tag_comb_dict = tag_comb_freq(filter_chemistry)
sorted_tag_comb = sorted(tag_comb_dict.items(), key=operator.itemgetter(1), reverse = True)
sorted_tag_comb
```

```python
total = chemistry.shape[0]
total
```

```python
tag_pair = list(map(lambda x: tuple([x[0], round(x[1]/total, 4)]), sorted_tag_comb))
tag_pair
```

```python
tag_comb_prob = {k: v / total for k, v in tag_comb_dict.items()}
tag_comb_prob
```

```python
def tag_freq(subject):
    subj_dict = {}
    for tags in subject["Tag_list"]:
        for tag in tags:
            if tag not in subj_dict:
                subj_dict[tag] = 1
            else:
                subj_dict[tag] += 1
    return subj_dict
```

```python
tag_dict = tag_freq(chemistry)
sorted_tag = sorted(tag_dict.items(), key=operator.itemgetter(1), reverse = True)
sorted_tag
```

```python
tag = list(map(lambda x: tuple([x[0], round(x[1]/total, 4)]), sorted_tag))
tag
```

```python
tag_prob = {k: v / total for k, v in tag_dict.items()}
```

```python
tag_prob
```

```python
import math
```

# coef(X,Y) = (E(XY) - E(X)E(Y)) / (sqrt(E(X^2) - (E(X))^2) * sqrt(E(Y^2) - (E(Y))^2))

```python
tag_comb_coef = {}
for k, v in tag_comb_prob.items():
    p_a = tag_prob[k[0]]
    p_b = tag_prob[k[1]]
    p_ab = v
    coef = (p_ab - p_a * p_b) / (math.sqrt(p_a - p_a*p_a) * math.sqrt(p_b - p_b*p_b))
    tag_comb_coef[k] = coef
```

```python
tag_comb_coef
sorted_tag_comb_coef = sorted(tag_comb_coef.items(), key=operator.itemgetter(1), reverse = True)
sorted_tag_comb_coef
```

# p(x|y) = p(xy) / p(y)

```python
tag_comb_cond_prob = {}
for k, v in tag_comb_prob.items():
    p_a = tag_prob[k[0]]
    p_b = tag_prob[k[1]]
    p_ab = v
    p_cond_b = p_ab / p_b
    p_cond_a = p_ab / p_a
    tag_comb_cond_prob[k] = tuple([p_cond_b, p_cond_a])
```

```python
tag_comb_cond_prob
```

```python
sorted_tag_comb_cond_prob = sorted(tag_comb_cond_prob.items(), key = lambda tup: tup[1][0], reverse = True)
sorted_tag_comb_cond_prob
```

```python
sorted_tag_comb_cond_prob = sorted(tag_comb_cond_prob.items(), key = lambda tup: tup[1][1], reverse = True)
sorted_tag_comb_cond_prob
```

# filter the tag pairs whose conditional probability satisfy: p(a|b) similar to p(b|a)

```python
filter_tag_comb_cond_prob = dict(filter(lambda elem: abs(elem[1][0] - elem[1][1]) < 0.5, tag_comb_cond_prob.items()))
filter_tag_comb_cond_prob
```

```python
sorted_filter_tag_comb_cond_prob = sorted(filter_tag_comb_cond_prob.items(), key = lambda tup: tup[1][0], reverse = True)
sorted_filter_tag_comb_cond_prob
```

```python
filter_tag_comb_cond_prob = dict(filter(lambda elem: abs(elem[1][0] - elem[1][1]) < 0.5 and elem[1][0] > 0.2 and elem[1][1] > 0.2, tag_comb_cond_prob.items()))
filter_tag_comb_cond_prob
```
